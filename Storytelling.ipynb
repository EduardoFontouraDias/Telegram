<h1 style="font-size: 120%;">História do Projeto: Chatbots e Pipeline de Dados via Telegram </h1>

Há alguns anos, a ascensão dos chatbots revolucionou a forma como interagimos digitalmente. Esses assistentes virtuais automatizados oferecem uma maneira eficiente e acessível de comunicação, e uma plataforma que se destaca nesse cenário é o Telegram, com sua API de bots poderosa e flexível.

No entanto, conforme a utilização de chatbots cresceu, percebemos a importância de diferenciar entre dados transacionais e analíticos. Os dados transacionais, gerados durante as interações em tempo real, fornecem informações cruciais para melhorar a experiência do usuário. Por outro lado, os dados analíticos oferecem insights valiosos que podem orientar estratégias de longo prazo e tomada de decisões.

Para preencher essa lacuna, nasceu o projeto de um pipeline de dados via Telegram. Vamos explorar sua arquitetura e como ele proporciona uma abordagem abrangente para lidar com ambos os tipos de dados.

**Arquitetura do Pipeline de Dados**

image.png



**Sistema Transacional: Telegram e sua API de Bots**

O projeto começa com a compreensão do Telegram, uma plataforma de mensagens poderosa. Utilizando sua API de Bots, podemos criar um chatbot capaz de receber mensagens em um grupo específico. Cada interação é então capturada em tempo real.

**Sistema Analítico: Ingestão, ETL e Apresentação com Códigos**

A transformação desses dados transacionais em analíticos ocorre em três etapas essenciais:

**Ingestão:** As mensagens do grupo do Telegram são capturadas via API GetUpdates e encaminhadas para a AWS através do API Gateway.

**ETL (Extração, Transformação e Carga)**

Os dados crus são armazenados em um bucket da AWS no formato JSON e, em seguida, passam por um processo de transformação em um formato mais eficiente, como o Parquet, utilizando funções Lambda.

image.png

Os dados transformados são organizados e armazenados em um novo bucket. E a partir disso é criado uma tabela particionada chamada "telegram" no AWS Athena, permitindo consultas eficientes.

**Análise Exploratória de Dados**

**Fonte:** Utilizando o método getUpdates da API do Telegram, analisamos os dados assim que são produzidos, proporcionando uma visão em tempo real das interações.

**Destino:** Na fase de apresentação, realizamos consultas SQL no AWS Athena para explorar os dados, identificando padrões, tendências e oportunidades de melhoria.

**Infraestrutura: Atualização Automática de Partições no AWS Athena**

Uma melhoria significativa seria implementar a atualização automática de novas partições na tabela do AWS Athena. Isso garantiria maior eficiência e agilidade no acesso aos dados mais recentes. Para isso foi utilizado uma nova função Lambda integrada com o Event Bridge que foi customizado para atualizar 1 hora após a ingestão e tratamento dos dados pelo AWS API Getway.

**Conclusão**

O projeto de pipeline de dados via Telegram é um projeto emocionante na integração de chatbots com análises de dados robustas. Ao utilizar a arquitetura proposta e implementar melhorias contínuas, buscamos aprimorar a experiência do usuário e fornecer informações estratégicas fundamentais para o crescimento e desenvolvimento contínuos.
